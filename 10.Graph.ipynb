# グラフ: Graph
- グラフは要素の関係を視覚的に表す
- 頂点 (ノード) と辺 (エッジ) から構成される: $G=(V, E)$
- 無向グラフ: 辺に方向がないグラフ
- 有向グラフ: 辺に方向があるグラフ
- 辺に数値が割り当てられている場合にネットワーク (network) と呼ぶこともある

----
- A graph is a visual representation of relationships between elements
- It consists of vertices (nodes) and edges: $G=(V, E)$
- Undirected graph: A graph where edges have no direction
- Directed graph: A graph where edges have direction
- A graph with numerical values assigned to edges is sometimes called a network

## グラフを定義する: Defining Graphs
- `networkx` ライブラリを使用してグラフを定義
    - `nx.Graph()` で無向グラフを作成
    - `nx.DiGraph()` で有向グラフを作成
- 頂点は整数または文字列で表す
- `add_edge` メソッドを使用して辺を追加
‐ 辺をタプルで定義し、`add_edges_from` メソッドを使用して複数の辺をリストにして一度に追加
- 頂点の位置を指定してグラフを描画することも可能
----
- Using the `networkx` library to define graphs
    - Create an undirected graph with `nx.Graph()`
    - Create a directed graph with `nx.DiGraph()`
- Vertices can be represented as integers or strings
- Use the `add_edge` method to add edges
- Define edges as tuples and use the `add_edges_from` method to add multiple edges at once from a list
- It is also possible to specify the position of vertices when drawing the graph

----

## グラフを描く: Drawing Graphs
- `nx.draw()` 関数を使用してグラフを描画
- `with_labels=True` を指定して頂点のラベルを表示
- 頂点の位置を指定してグラフを描画することも可能
----
- Use the `nx.draw()` function to draw graphs
- Specify `with_labels=True` to display vertex labels
- It is also possible to specify the position of vertices when drawing the graph



```python
def define_graph():
    """
    頂点を文字列で表す無向グラフ
    """
    G = nx.Graph()
    G.add_edge("a", "b")
    G.add_edge("a", "c")
    G.add_edge("b", "d")
    G.add_edge("c", "d")
    G.add_edge("d", "e")
    G.add_node("f")  # 孤立した頂点の追加
    return G
```

```python
g = define_graph()
nx.draw(g, with_labels=True)
```

```python
def define_graph():
    """
    頂点を整数で表す無向グラフ
    頂点の位置も指定
    """
    G = nx.Graph()  # 無向グラフの作成: Create an undirected graph
    # 辺の追加: Add edges
    edges = [
        (1, 2),
        (1, 3),
        (2, 3),
        (2, 5),
        (2, 4),
        (3, 5),
        (3, 4),
        (5, 4),
        (4, 6),
        (5, 6),
    ]
    G.add_edges_from(edges)
    # 頂点の位置を設定: オプション
    # Set node positions (optional)
    positions = {
        1: (0, 0),
        2: (1, 1),
        3: (1, -1),
        4: (2, 1),
        5: (2, -1),
        6: (3, 0),
    }
    nx.set_node_attributes(G, positions, "pos")
    return G
```

```python
g = define_graph()
pos = nx.get_node_attributes(g, "pos")
nx.draw(g, pos=pos, with_labels=True, node_color="lightblue", edge_color="black")
```

```python
def define_directed_graph():
    G = nx.DiGraph()  # 有向グラフの作成: Create a directed graph
    # 辺の追加: Add edges
    edges = [
        (1, 2),
        (1, 3),
        (2, 4),
        (2, 5),
        (3, 4),
        (3, 6),
        (4, 5),
        (5, 7),
        (6, 5),
        (7, 6),
    ]
    G.add_edges_from(edges)
    # 頂点の位置を設定: オプション
    # Set node positions (optional)
    positions = {
        1: (0, 0),
        2: (1, 1),
        3: (1, -1),
        4: (1, 0),
        5: (2, 0),
        6: (3, -1),
        7: (3, 1),
    }
    nx.set_node_attributes(G, positions, "pos")
    return G
```

```python
g_d = define_directed_graph()
pos = nx.get_node_attributes(g_d, "pos")
nx.draw(g_d, pos=pos, with_labels=True, node_color="lightblue", edge_color="black")
```

## Euler閉路を列挙: Enumerating Eulerian Circuits

```python
def enumerate_Euler_main(G:nx.Graph, s:int|str)->list[tuple[int|str,int|str]]:
    """
    Euler閉路を列挙する関数: Function to enumerate Eulerian circuits:
    """
    G_copy = G.copy()  # グラフのコピーを作成: Create a copy of the graph
    euler_paths = []
    
    def find_euler_path(v:int|str, path:list[tuple[int|str,int|str]], ss:int|str):
        # 終了条件: 全ての辺を訪れ、開始点に戻る: Base case: All edges visited and back to start
        if len(path) == G_copy.number_of_edges() and v == ss:
            euler_paths.append(path.copy())
            return
        
        for neighbor in list(G_copy.neighbors(v)):# 隣接する頂点を探索: Explore adjacent vertices
            if (v, neighbor) not in path and (neighbor, v) not in path:
                # まだ訪れていない辺を選択: Choose an edge that has not been visited
                path.append((v, neighbor))
                find_euler_path(neighbor, path, ss)
                path.pop()
    
    find_euler_path(s, [], s)
    return euler_paths

def euler_example():
    """ 
    Eulerを列挙するグラフの例: Example graph for enumerating Eulerian circuits
    """
    G = nx.Graph()
    G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4), (2, 5), (3, 4), (3, 5)])
    pos = {1: (0, 0), 2: (-1, -1), 3: (1, -1), 4: (-1, -2), 5: (1, -2)}
    nx.set_node_attributes(G, pos, "pos")
    return G

g = euler_example()
pos = nx.get_node_attributes(g, "pos")
nx.draw(g, with_labels=True, node_color="lightblue", edge_color="black", pos=pos)
paths = enumerate_Euler_main(g, 1)
print("Eulerian Circuits:", len(paths))
for path in paths:
    print(path)
```
